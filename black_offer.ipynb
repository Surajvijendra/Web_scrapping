```python
# importing the necesaary libraries
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

import ast

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# importing the data
data = pd.read_excel(r'/content/Input.xlsx')
data
```

```python
# importing the articles from the corresponding url's
titles = []
articles = []
url_ids = []
titile_article = []
URL = []

for i in range(len(data['URL'])):
    url = data['URL'][i]
    url_id_value = data['URL_ID'][i]
    URL.append(url)
    try:
        req = requests.get(url)
        status = req.status_code
        if status == 200:
            soup = BeautifulSoup(req.text, 'lxml')

            # Extracting the title
            title_tag = soup.find('h1',class_='entry-title')
            title_tag_1 = soup.find('h1',class_='tdb-title-text')
            if title_tag:
              title_text = title_tag.get_text()
            elif title_tag_1:
              title_text = title_tag_1.get_text()
            else:
              title_text = 'No title found'

            # Extracting the article

            div = soup.find('div',class_='td-post-content tagdiv-type')
            div_s = soup.find_all('div',class_='tdb-block-inner td-fix-index')
            if div:
              div_text = div.get_text()


            elif div_s :
              if len(div_s) >= 10:
                div_s = div_s[14]
                div_text = div_s.get_text()

            else:
              div_text = 'No content found'


            # Combining title and article
            full_text = f'Title: {title_text}\nArticle: {div_text}'
            titles.append(title_text)
            articles.append(div_text)
            titile_article.append(full_text)
            url_ids.append(url_id_value)

            # Save title and article
        else:
          titles.append(f"HTTP Error: {status}")
          articles.append(f"HTTP Error: {status}")
          titile_article.append(f"HTTP Error: {status}")
          url_ids.append(url_id_value)

    except Exception as e:
        titles.append(f'Error: {e}')
        articles.append("")
        url_ids.append(url_id_value)
```

```python
print(f'extracted titles : {titles}')
```

```python
print(f'extracted articles : {articles}')
```

```python
articles = [sentence_in_string.replace("\n", "") for sentence_in_string in articles]
```

data cleaning using stopwords

```python
stopwords_list = []
path = ['/content/StopWords_Auditor.txt','/content/StopWords_Currencies.txt','/content/StopWords_DatesandNumbers.txt',
        '/content/StopWords_Generic.txt','/content/StopWords_GenericLong.txt','/content/StopWords_Geographic.txt',
        '/content/StopWords_Names.txt']

for path_list in range(len(path)):
  with open(path[path_list],'r', encoding='latin-1') as file:
    stopwords_1 = file.read().split()
    stopwords_list.extend(stopwords_1)
```

```python
print(f'stopwords list : {str(stopwords_list).lower()}')
```

```python
lemma = WordNetLemmatizer()
```

```python
print(f'list of sentences before cleaning : {sentences_list}')
```

```python
# removing the stopwords
text_sentences = []
list_of_words = []
for sentences_for_cleaning in range(len(sentences_list)):
  text = re.sub('[^a-zA-Z.]',' ',str(sentences_list[sentences_for_cleaning]))
  text = text.lower()
  sentences = nltk.sent_tokenize(text)
  words_articles = nltk.word_tokenize(str(sentences))
  words_articles = [lemma.lemmatize(word) for word in words_articles if not word in set(stopwords_list)]
  articles_words_join = ' '.join(words_articles)
  text_sentences.append(articles_words_join)
  list_of_words.append(words_articles)
```

```python
print(f'list of sentences after cleaning : {text_sentences}')
```

```python
article_sentences = []
for artic in text_sentences:
  list_of_sentences = ast.literal_eval(artic)
  article_sentences.append(list_of_sentences)
```

```python
article_sentences[0][0]
```

```python
print(f'list of words{list_of_words[0]}')
```

```python
# removing the extra spaces and other special charecters
cleaned_words_list = []
unwanted = re.compile(r'\w+')
for cleaning_word in range(len(list_of_words)):
  cleaned_words = unwanted.findall(str(list_of_words[cleaning_word]))
  cleaned_words_list.append(cleaned_words)
```

```python
print(f'list of words after cleaning 1st row : {cleaned_words_list[0]}')
```

```python
df = pd.DataFrame({'url_id':url_ids,'url':URL,'article_text':article_sentences,'word_text':cleaned_words_list})
df.head()
```

```python
# creating the positive words list
positive_words = []
positive_words_path = '/content/positive-words.txt'
with open(positive_words_path, 'r',encoding='latin-1') as file:
    positive = file.read().split()
    positive_words.extend(positive)
```

```python
print(f'positive words are : {positive_words}')
```

```python
# creating the negative words list
negative_words = []
negative_words_path = '/content/negative-words.txt'
with open(negative_words_path, 'r',encoding='latin-1') as file:
    negative = file.read().split()
    negative_words.extend(negative)
```

```python
print(f'negative words are : {negative_words}')
```

```python
# calculating +ve and -ve words
def calculate_score(text):
    positive_score = 0
    negative_score = 0
    for word in text:
        if word in positive_words:
            positive_score += 1
        elif word in negative_words:
            negative_score -= 1
    return positive_score, negative_score
```

```python
positive_scores = []
negative_scores = []

for text in df['word_text']:
  pos_score, neg_score = calculate_score(text)
  positive_scores.append(pos_score)
  negative_scores.append(neg_score * (-1))
```

```python
print(f'positive score list {positive_scores}')
```

```python
print(f'negative score list {negative_scores}')
```

```python
df['positive_score'] = positive_scores
df['negative_score'] = negative_scores
df
```

```python
# calculating the polarity score
polarity_score = [(df['positive_score'][polarity] - df['negative_score'][polarity])/((df['positive_score'][polarity] + df['negative_score'][polarity] + 0.000001)) for polarity in range(len(df['positive_score']))]
```

```python
polarity_score = np.array(polarity_score)
polarity_score = np.round(polarity_score,2)
df['polarity_score'] = polarity_score
df.head(10)
```

```python
# calculating the subjective score
subjective_score = [(positive_scores[subjective]+negative_scores[subjective])/(len(df['word_text'][subjective])+0.000001) for subjective in range(len(df['word_text']))]
```

```python
subjective_score = np.array(subjective_score)
subjective_score = np.round(subjective_score,2)
df['subjective_score'] = subjective_score
```

```python
# set of vowes
vowels = set('aeiou')
```

```python
# finding the syllbels
syllabel_count = []
for para in df['word_text']:
  syllabel_count_para = []
  for word in para:
    syllabel_per_word = 0
    for letter in word:
      if letter in vowels:
        syllabel_per_word += 1
    if word.endswith('es') or word.endswith('ed'):
      syllabel_per_word -= 1
    if syllabel_per_word <= 0:
      syllabel_per_word = 0
    syllabel_count_para.append(syllabel_per_word)
  syllabel_count.append(syllabel_count_para)
```

```python
syllabel_count[0][0]
```

```python
df['syllable_per_word'] = syllabel_count
```

```python
# calculating the complex words
complex_word_count = []
for complex_word_in_para in syllabel_count:
  complex_word = 0
  for word_list in complex_word_in_para:
    if word_list >= 2:
      complex_word += 1
  complex_word_count.append(complex_word)
```

```python
print(f'list of complex word count{complex_word_count}')
```

```python
df['complex_word_count'] = complex_word_count
df
```

```python
# calculating the average sentence length
average_sentence_length = [((len(df['word_text'][length]))/(len(df['article_text'][length]))) for length in range(len(df['article_text']))]
```

```python
average_sentence_length = np.array(average_sentence_length)
average_sentence_length = np.round(average_sentence_length,0)
df['average_sentence_length'] = (average_sentence_length)
```

```python
# calculating the percentage complex word
percentage_complex_word = [((df['complex_word_count'][percentage]))/(len(df['word_text'][percentage])) for percentage in range(len(df['complex_word_count']))]
```

```python
percentage_complex_word = np.array(percentage_complex_word)
percentage_complex_word = np.round(percentage_complex_word,2)
df['percentage_complex_word'] = percentage_complex_word
```

```python
# calaculating the fog index
fog_index = [(0.4*((df['average_sentence_length'][ind])+ (df['percentage_complex_word'][ind]))) for ind in range(len(df['url']))]
```

```python
fog_index = np.array(fog_index)
fog_index = np.round(fog_index,0)
df['fog_index'] = fog_index
```

```python
# calculating the word count
df['word_count'] = [len(df['word_text'][count]) for count in range(len(df['word_text']))]
```

```python
personal_pronouns = ['i','we','my','ours','us']
```

```python
# calculating the personal pronouns
def pronouns_count(personal_pronouns_text):
  pronouns_words = r'\b(i|we|my|ours|us)\b'
  pronouns = re.findall(pronouns_words,personal_pronouns_text)
  pronouns_count_score = 0
  for nouns in pronouns:
    if nouns == 'us':
      if not re.search(r'\bus\s+(country|of)',personal_pronouns_text):
        pronouns_count_score += 1
    else:
      pronouns_count_score += 1
  return pronouns_count_score
```

```python
personal_pronouns_count = []
for pronoun_word in df['article_text']:
  personal_pronouns_count.append(pronouns_count(str(pronoun_word)))
```

```python
# calculating the length of letters
def letters_count(list_of_words):
  letters_count_row_wise = []
  for letters in list_of_words:
    letters_count = sum(len(letter_list) for letter_list in letters)
    letters_count_row_wise.append(letters_count)
  return letters_count_row_wise
```
